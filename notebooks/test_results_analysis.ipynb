```python
from pathlib import Path
import pandas as pd
import os
```

```python
exp_folder = Path("../experiment_results")
list(exp_folder.iterdir())
```

```python
test_results_path = exp_folder / "test_results.csv"
test_results = pd.read_csv(test_results_path)
test_results
```

```python
test_results.columns
```

```python
#test_results = test_results.loc[~test_results["test/accuracy"].isna()].copy()
test_results.dropna(subset=["test/accuracy"], inplace=True)
test_results.loc[:,"weights"] = test_results["binary weights"].apply(lambda x: "binary weights" if x else "f32 weights")
test_results.loc[:,"activations"] = test_results["binary activations"].apply(lambda x: "binary activations" if x else "f32 activations")
test_results.loc[:,"algorithm"] = test_results["algorithm"].apply(lambda x: {
    "bp": "BP",
    "dfa": "DFA",
    "drtp": "DRTP",
    "hsic": "HSIC",
    "pepita": "SigpropTL",
}[x])
test_results.loc[:,"skip connections"] = test_results["removed skip connexions"].apply(lambda x: "present" if x == False else "removed")
```

```python
test_results
```

```python
best_acc = test_results.groupby(["id",
                                 "host",
                                 "gpu",
                                 "algorithm",
                                 "model",
                                 "binary activations",
                                 "binary weights",
                                 "weights",
                                 "activations",
                                 "removed skip connexions"], as_index=False)\
    .agg(accuracy=("test/accuracy", "max"),
         runtime=("_runtime", "max"),
         gpu_watts=("gpu_watts", "max"),
         gpu_total_mem=("gpu_total_mem", "max"),
         gpu_mem_alloc_percent=("gpu_mem_alloc_percent", "max"))
best_acc.loc[:, "gpu_mem_alloc"] = best_acc["gpu_mem_alloc_percent"] * best_acc["gpu_total_mem"] / 100.0
best_acc.loc[:, "gpu_mem_alloc (MB)"] = best_acc["gpu_mem_alloc"] / (10.0 ** 6)
best_acc.loc[:, "gpu_mem_alloc (GB)"] = best_acc["gpu_mem_alloc"] / (10.0 ** 9)
best_acc.loc[:, "accuracy"] = best_acc["accuracy"] * 100
```

```python
best_acc
```

```python
set(best_acc["gpu"]), set(best_acc["host"]), set(best_acc["gpu"] + " " + best_acc["host"])
```

```python
test_results = best_acc.groupby(["algorithm",
                                 "model",
                                 "binary activations",
                                 "binary weights",
                                 "removed skip connexions"], as_index=False)\
    .agg(accuracy=("accuracy", "mean"),
         accuracy_std=("accuracy", "std"))
test_results
```

#### Runtimes

```python
system = best_acc.groupby(["gpu",
                                 "algorithm",
                                 "model",
                                 "binary activations",
                                 "binary weights",
                                 "weights",
                                 "activations",
                                 "removed skip connexions"], as_index=False)\
    .agg(runtime=("runtime", "mean"),
         runtime_std=("runtime", "std"),
         watts=("gpu_watts", "mean"),
         watts_std=("gpu_watts", "std"),
         gpu_mem=("gpu_mem_alloc (GB)", "mean"),
         gpu_mem_std=("gpu_mem_alloc (GB)", "std"),)
system.loc[:, "Runtime (s)"] = [f"{acc:.0f} ({acc_std:.0f})" for acc, acc_std in zip(system["runtime"], system["runtime_std"])]
system.loc[:, "GPU memory"] = [f"{acc:.1f} ({acc_std:.1f})" for acc, acc_std in zip(system["gpu_mem"], system["gpu_mem_std"])]
system
```

```python
binary_system = system[(system["weights"] == "binary weights") & (system["activations"] == "binary activations")]
```

```python
binary_system.pivot_table(values="Runtime (s)", index=["model", "gpu"], columns=["algorithm"], aggfunc="first")
```

```python
binary_system
```

```python
binary_system.pivot_table(values="GPU memory", index=["model", "gpu"], columns=["algorithm"], aggfunc="first")
```

```python
system_binary_mem_rt = binary_system.pivot_table(values=["GPU memory", "Runtime (s)"], index=["gpu", "model"], columns=["algorithm"], aggfunc="first")
```

```python
system_binary_mem_rt_latex = system_binary_mem_rt.stack(0).reorder_levels((2, 0, 1)).sort_index(level=0)
system_binary_mem_rt_latex
```
